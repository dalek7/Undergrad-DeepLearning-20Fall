## 간단한 Forward 연산 해보기
* bias는 일단 무시하고 진행
### 입력 층 (input layer)

````python
X = np.array([1.0, 0.5])
````

````python
X
````

### 입력의 2개가 다음 layer에 있는 3개와 연결

````python
W1 = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
````

````python
W1.shape
````

````python
W1
````

### 곱해보기 (dot-product)

````python
a1 = np.dot(X, W1)
````
