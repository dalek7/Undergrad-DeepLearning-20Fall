분류 (classification) 문제의 경우 가장 큰 값을 선택하면 되기는 하지만...

즉, 위의 경우 class 가 2. 숫자를 해석하기가 애매함

Motivation: 위의 값을 확률처럼 해석하면 좋을텐데 ?! 

### Softmax의 등장  

```python
def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```python
y = softmax(a)
```

```python
print(y)
```

```python
np.sum(y)
```

* 확률처럼 더하면 1이 되고, 모두가 양수로 변환되었음을 확인 !
