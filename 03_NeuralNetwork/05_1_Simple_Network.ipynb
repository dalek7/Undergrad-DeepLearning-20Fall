```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a-c)   #ê°œì„ 
    sum_exp_a = sum(np.exp(a-c))
    y = exp_a/sum_exp_a
    return y
```

```python
def cross_entropy_error(y, t):
    if y.ndim == 1:
        t = t.reshape(1, t.size)
        y = y.reshape(1, y.size)
        
    # í›ˆë ¨ ë°ì´í„°ê°€ ONE-HOT ë²¡í„°ë¼ë©´ ì •ë‹µ ë ˆì´ë¸”ì˜ ì¸ë±ìŠ¤ë¡œ ë°˜í™˜
    if t.size == y.size:
        t = t.argmax(axis=1)
             
    batch_size = y.shape[0]
    return -np.sum(np.log(y[np.arange(batch_size), t] + 1e-7)) / batch_size
```

```python
class simpleNet:
    def __init__(self):
        np.random.seed(5) #í•­ìƒ ë™ì¼í•œ ëœë¤ê°’ ì¶œë ¥ì„ ìœ„í•¨ (ìˆ˜ì—… ëª©ì )
        self.W = np.random.randn(2,3) # ì •ê·œë¶„í¬ë¡œ ì´ˆê¸°í™”

    def predict(self, x):
        return np.dot(x, self.W)

    def loss(self, x, t):
        z = self.predict(x)
        y = softmax(z)
        loss = cross_entropy_error(y, t)

        return loss
```

```python
net = simpleNet()
```

```python
x = np.array([0.6, 0.9]) # ì…ë ¥
t = np.array([0, 0, 1]) #ì •ë‹µ, TARGET
```
