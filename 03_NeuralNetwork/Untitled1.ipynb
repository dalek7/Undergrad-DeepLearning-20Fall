```python
import numpy as np
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def _numerical_gradient_no_batch(f, x):
    h = 1e-4 # 0.0001
    #print('x.shape={}'.format(x.shape)) # (324,)
    #print('x.size={}'.format(x.size)) #324
    grad = np.zeros_like(x) # x와 형상이 같은 배열을 생성
    
    for idx in range(x.size):
        tmp_val = x[idx] #single_nㅕㅡㅠㄷㄱ
        
        # f(x+h) 계산
        x[idx] = float(tmp_val) + h
        fxh1 = f(x)
        
        # f(x-h) 계산
        x[idx] = tmp_val - h 
        fxh2 = f(x) 
        
        print(idx, '-->', fxh1, '\t', fxh2)
        grad[idx] = (fxh1 - fxh2) / (2*h)
        x[idx] = tmp_val # 값 복원
        
    return grad
```

```python
def numerical_gradient(f, X):
    if X.ndim == 1:
        return _numerical_gradient_no_batch(f, X)
    else:
        grad = np.zeros_like(X)
        
        for idx, x in enumerate(X):
            grad[idx] = _numerical_gradient_no_batch(f, x)
        
        return grad    
```

```python
def function_2(x):
    if x.ndim == 1:
        return np.sum(x**2)
    else:
        return np.sum(x**2, axis=1)
```

```python
x0 = np.arange(-2, 2.5, 0.25)
x1 = np.arange(-2, 2.5, 0.25)
X, Y = np.meshgrid(x0, x1)
print(X.shape,Y.shape)
```

```python
X = X.flatten()
Y = Y.flatten()

print(X.shape, np.array([X, Y]).shape, np.array([X, Y]).ndim)
```

```python
arr = np.array([X, Y])
print(arr.shape, arr.ndim)
```

```python
grad = numerical_gradient(function_2, arr )
print(grad.shape)
```
